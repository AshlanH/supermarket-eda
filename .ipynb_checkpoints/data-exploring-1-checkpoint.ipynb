```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("supermarket_sales.csv")
```

```python
#create datettime column
df['DateTime'] = df['Date'] + ' ' + df['Time']
df['DateTime'] = df['DateTime'].astype('datetime64[ns]')
df.drop(['Date','Time'], inplace=True, axis=1)
df['Month'] = df['DateTime'].apply(lambda x: x.month)
df['Day'] = df['DateTime'].apply(lambda x: x.day)
```

### Distribution of product lines across different branches

```python
g = sns.displot(
    data=df,
    #hue = 'Branch',
    x = 'Product line',
    kind = 'hist',
    aspect = 1,
    multiple = 'stack',
    height = 8,
    palette = 'crest'
)

g.set_xticklabels(rotation=45, fontsize = 10)
g.ax.set_ylabel('Total')
g.ax.set_title('Distribution of product lines across different branches')
plt.show()
```

```python
df.groupby(['Month', 'Day'])[['Total']].count()
```

```python
sns.displot(
    df.groupby(['Month', 'Day'])[['Total']].count(),
    x = 'Day',
    y = 'Total',
    hue = 'Month',
    bins=30
)
```

```python
df.groupby(['Month', 'Day'])[['Total']].sum()
```

```python
g = sns.relplot(
    data=df.groupby(['Month', 'Day'])[['Total']].sum(),
    hue = 'Month',
    x = 'Day',
    y = 'Total',
    kind = 'line',
    lw = 2.5,
    aspect = 2,
    height = 8
)

g.ax.set_xticks([i for i in range (1, 31)])
plt.show()
```
